```python
import pandas as pd
import sqlite3

def load_data_to_database():
    # Leer el archivo CSV y excluir la primera columna sin nombre
    df = pd.read_csv('./data/Advertising.csv', usecols=lambda column: column != 'Unnamed: 0')

    # Conectar a la base de datos SQLite
    connection = sqlite3.connect('./data/advertising_data.db')

    # Guardar el DataFrame en la tabla "advertising_data" en la base de datos
    df.to_sql('advertising_data', connection, index=False, if_exists='append')

    # Cerrar la conexión con la base de datos
    connection.close()

load_data_to_database()

```
